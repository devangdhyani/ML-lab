```python
import pandas as pd
import numpy as np
df =pd.read_excel("Student-List-Exp.xlsx")
df.head()
```

```python
df.rename(columns={"Branch /Course":"Course"},inplace=True)
df.rename(columns={"University Registration No. (Issued by R&S Branch)":"Registration No"},inplace=True)
df.rename(columns={"Date of Admission in the Present Class/Course":"Admission date"},inplace=True)
df.rename(columns={"Aadhaar No. of Student":"Adhaar no"},inplace=True)
df.rename(columns={"Spelling strictly as per Registration record in Capital Letters (Issued by R&S Branch)-SN":"Student name"},inplace=True)
df.rename(columns={"Spelling strictly as per Registration record in Capital Letters (Issued by R&S Branch)-FN":"Father name"},inplace=True)
df.rename(columns={"Updated Mobile No. of Student":"Mobile No"},inplace=True)
df.rename(columns={"Updated E-mail ID":"Email ID"},inplace=True)
df.rename(columns={"Total Due Fee for Current Semester / (Previous semester if pending)":"Due Fee Pending"},inplace=True)
                        
df.head()
```

```python
print("\n List NUMBER of DUES ")
print(df[df["Due Fee Pending"]!=0].loc[:,"Student name"])
```

```python
#List of all students where mentor name is Dr INDU KASHYAP
print("Students with Dr INDU KASHYAP:-\n")
print(df[df["mentors"]=="Dr INDU KASHYAP"].loc[:,"Student name"])
```

```python

#List of Students where Aadhaar No. is not available
print("Students with missing Aadhaar No.:-\n")
print(df[df["Adhaar no"]==np.nan].loc[:,"Student name"])
     
```

```python
print("LIST of student from Delhi :-\n")
print(df[df["State"]=="DELHI"].loc[:,"Student name"])
```
